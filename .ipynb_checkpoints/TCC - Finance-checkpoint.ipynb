# Projeto de Conclusão de Curso
## Objetivo
> Construir um modelo de machine learning que possibilite assistência na compra ou venda de ações

## 1. Importações e coleta de dados
> Importação das bibliotecas à serem utilizadas e obtenção dos dados que irão compor os datasets iniciais

### 1.1 Preço das ações

```python
import pandas as pd
import os
from pandas_datareader import data as web
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import seaborn as sns

empresas = ["ABEV3", "AZUL4", "BTOW3", "B3SA3", "BBSE3", "BRML3", "BBDC4", "BRAP4", "BBAS3", "BRKM5", "BRFS3", "BPAC11", "CRFB3", "CCRO3", "CMIG4", "HGTX3", "CIEL3", "COGN3", "CPLE6", "CSAN3", "CPFE3", "CVCB3", "CYRE3", "ECOR3", "ELET6", "EMBR3", "ENBR3", "ENGI11", "ENEV3", "EGIE3", "EQTL3", "EZTC3", "FLRY3", "GGBR4", "GOAU4", "GOLL4", "NTCO3", "HAPV3", "HYPE3", "IGTA3", "GNDI3", "ITSA4", "ITUB4", "JBSS3", "JHSF3", "KLBN11", "RENT3", "LCAM3", "LAME4", "LREN3", "MGLU3", "MRFG3", "BEEF3", "MRVE3", "MULT3", "PCAR3", "PETR4", "BRDT3", "PRIO3", "QUAL3", "RADL3", "RAIL3", "SBSP3", "SANB11", "CSNA3", "SULA11", "SUZB3", "TAEE11", "VIVT3", "TIMS3", "TOTS3", "UGPA3", "USIM5", "VALE3", "VVAR3", "WEGE3", "YDUQ3"]

cotacao = pd.DataFrame(columns = ["Date","High","Low","Open","Close","Volume","Adj Close","Empresa"])

for empresa in empresas:
    try:
        cotacao_append = web.DataReader(empresa + ".SA", data_source="yahoo", start="01-01-2010", end="31-12-2022")
        cotacao_append["Empresa"] = empresa
        cotacao_append.insert(0, 'Date', cotacao_append.index)
        cotacao_append = cotacao_append.reset_index(drop=True)
        # loop para adicionar cotação dia 31/12 igual ao dia anterior
        for i in list(pd.DatetimeIndex(cotacao_append['Date']).year.unique()):
            cotacao_append = cotacao_append.sort_values(by='Date')
            cotacao_append = cotacao_append.reset_index(drop=True)
            date_loc = cotacao_append.loc[cotacao_append["Date"] <= str(i) + "-12-31"].iloc[-1]
            date_loc["Date"] = str(i) + "-12-31 00:00:00"
            date_loc["Date"] = pd.to_datetime(date_loc["Date"] )
            cotacao_append = cotacao_append.append(date_loc)
        
        
        cotacao = pd.concat([cotacao_append, cotacao], ignore_index=True)
        print(f"Importação de dados da empresa {empresa} Concluída")
    except:
        print(f"Importação de dados da empresa {empresa} Falhou")
```

### 1.2 Coletar balanços das empresas

```python
fundamentos = {}
for i in os.listdir("balancos"):
    nome = i[:-4]
    if nome in empresas:
        balanco = pd.read_excel("balancos/" + i, sheet_name = 0)
        #colocando o nome da empresa no cabeçalho
        balanco.iloc[0,0] = nome
        balanco.columns = balanco.iloc[0]
        balanco = balanco[1:] #dropando linha repetida
        balanco = balanco.set_index(nome)
        
        
        
        
        dre = pd.read_excel(r"balancos/" + i, sheet_name = 1)
        dre.iloc[0,0] = nome
        dre.columns = dre.iloc[0]
        dre = dre[1:] #dropando linha repetida
        dre = dre.set_index(nome)
        fundamentos[nome]=pd.concat([balanco,dre])
        
        
```

### 1.3 Tratamento

```python
#dicionário
cotacoes_dict = {}
for i in cotacao["Empresa"].unique():
    cotacoes_dict[i] = cotacao.loc[cotacao["Empresa"]==i, :]
```

```python
#cotacao.to_excel(r"C:\Users\Franc\Downloads\output234234.xlsx")
#cotacao = pd.read_excel(r"C:\Users\Franc\Downloads\output234234.xlsx")
```

```python
#not necessary? - remover cotações nulas
for i in cotacoes_dict:
    if cotacoes_dict[i].isnull().values.any():
        cotacoes_dict.pop(i)
        fundamentos.pop(i)

empresas = list(cotacoes_dict.keys())
```

```python
#juntar fundamentos e cotação
for empresa in fundamentos:
    tabela = fundamentos[empresa].T
    tabela.index = pd.to_datetime(tabela.index, format = "%d/%m/%Y")
    tabela_cotacao = cotacoes_dict[empresa].set_index("Date")
    tabela_cotacao = tabela_cotacao[["Adj Close"]]
    
    tabela = tabela.merge(tabela_cotacao,right_index=True, left_index = True)
    tabela.index.name = empresa
    fundamentos[empresa] = tabela
```

```python
#remoção de colunas diferentes entre as tabelas
#definir padrão
colunas = list(fundamentos["ABEV3"].columns)

for empresa in fundamentos.copy().keys():
    if set(colunas) != set(fundamentos[empresa].columns):
        fundamentos.pop(empresa)
```

```python
#adicionando nº nas colunas com mesmo nome - para diferenciar
texto_colunas = ";".join(colunas)

colunas_modificadas = []
for coluna in colunas:
    if colunas.count(coluna) == 2 and coluna not in colunas_modificadas:
        texto_colunas = texto_colunas.replace(";" + coluna + ";",";" + coluna + "_1;", 1)
        colunas_modificadas.append(coluna)
colunas = texto_colunas.split(';')

#inserindo colunas na tabela de fundamentos
for empresa in fundamentos:
    fundamentos[empresa].columns = colunas
```

```python
#Análise de linhas vazias
valores_vazios = dict.fromkeys(colunas, 0)
total_linhas = 0
for empresa in fundamentos:
    tabela = fundamentos[empresa]
    total_linhas += tabela.shape[0] #soma das linhas - todas as empresas
    for coluna in colunas:
        qtde_vazios = pd.isnull(tabela[coluna]).sum()
        valores_vazios[coluna] += qtde_vazios #soma de valores vazios na coluna - todas as empresas
print(f"Colunas e total de valores vazios:\n{valores_vazios}")
print(f"Total de linhas:{total_linhas}")
```

```python
#remover colunas que tem mais de x valores vazios
remover_colunas = []
for coluna in valores_vazios:
    if (valores_vazios[coluna]/total_linhas) > 0.25: # se a porcentagem de valores vazios for maior que 25% excluir
        remover_colunas.append(coluna)

for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa].drop(remover_colunas, axis=1)
    fundamentos[empresa] = fundamentos[empresa].fillna(method='bfill') #preencher valores vazios com valor mais próximo (acima)
```

## 2. Análise 
* Subiu mais do que o Ibovespa (ou caiu menos) -> Comprar (label = 2) --> Ex. Ação 10%, ibovespa 5%
* Diferença entre Ibovespa e ação até 2% -> Não Comprar (label = 1) --> Ex. Ação 10%, Ibovespa 11%
* Diferença entre Ibovespa e ação maior que 2% -> Vender (label = 0) --> Ex. Ação 10%, Ibovespa 20%

## 2.1 Inserção de labels

```python
#cotação Ibovespa
data_inicial = "12/30/2012"
data_final = "31/12/2022"

from pandas_datareader import data as web
df_ibov = web.DataReader('^BVSP', data_source='yahoo', start=data_inicial, end=data_final)
```

```python
#inserindo cotação do dia 31/12 (originalmente feriado)
df_ibov.insert(0, 'Date', df_ibov.index)
df_ibov = df_ibov.reset_index(drop=True)
for i in list(pd.DatetimeIndex(df_ibov['Date']).year.unique()):
    df_ibov = df_ibov.sort_values(by='Date')
    df_ibov = df_ibov.reset_index(drop=True)
    date_loc = df_ibov.loc[df_ibov["Date"] <= str(i) + "-12-31"].iloc[-1]
    date_loc["Date"] = str(i) + "-12-31 00:00:00"
    date_loc["Date"] = pd.to_datetime(date_loc["Date"] )
    df_ibov = df_ibov.append(date_loc)

df_ibov = df_ibov.rename(columns={"Adj Close": "IBOV"})
```

```python
#inserindo cotação do ibov na tabela de fundamentos
df_ibov = df_ibov.set_index('Date')
for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa].merge(df_ibov[["IBOV"]], left_index=True, right_index=True)
    #fundamentos[empresa] = fundamentos[empresa].drop("IBOV_y", axis=1)
    #fundamentos[empresa] = fundamentos[empresa].rename(columns={"IBOV_x": "IBOV"})
display(fundamentos["ABEV3"].head(5))
```

```python
#tornar cada coluna na tabela em % - pegar variação entre trimestres
for empresa in fundamentos:
    tabela = fundamentos[empresa]
    tabela = tabela.sort_index()
    for coluna in tabela:
        if "Adj Close" in coluna or "IBOV" in coluna:
            pass
        else:
            condicoes = [(tabela[coluna].shift(1) == 0)]
            valores = [1]
            tabela[coluna] = np.select(condicoes, valores, default=(tabela[coluna] - tabela[coluna].shift(1)) / abs(tabela[coluna].shift(1)))
    #formato para adj close e ibov é diferente:
    tabela["Adj Close"] = tabela["Adj Close"].shift(-1) / tabela["Adj Close"] - 1
    tabela["IBOV"] = tabela["IBOV"].shift(-1) / tabela["IBOV"] - 1
    #colocar label de venda, compra ou não compra 

    tabela["Resultado"] = tabela["Adj Close"] - tabela["IBOV"]
    condicoes = [
        (tabela["Resultado"] > 0),
        (tabela["Resultado"] < 0) & (tabela["Resultado"] >= -0.02),
        (tabela["Resultado"] < -0.02)
    ]
    valores = [2, 1, 0]
    tabela["Decisao"] = np.select(condicoes, valores)
    
    fundamentos[empresa] = tabela
    
display(fundamentos["ABEV3"])
backup_dates = fundamentos.copy()
```

```python
#analise de valores vazios
colunas = list(fundamentos["ABEV3"].columns)
valores_vazios = dict.fromkeys(colunas, 0)
total_linhas = 0
for empresa in fundamentos:
    tabela = fundamentos[empresa]
    total_linhas += tabela.shape[0] #soma das linhas - todas as empresas
    for coluna in colunas:
        qtde_vazios = pd.isnull(tabela[coluna]).sum()
        valores_vazios[coluna] += qtde_vazios #soma de valores vazios na coluna - todas as empresas
print(f"Colunas e total de valores vazios:\n{valores_vazios}")
print(f"Total de linhas:{total_linhas}")
```

```python
#remover colunas que tem mais de x valores vazios
remover_colunas = ["Adj Close", "IBOV", "Resultado"] #remover estas para não deixar variáveis demais relacionadas
for coluna in valores_vazios:
    if (valores_vazios[coluna]/total_linhas) > 0.25: # se a porcentagem de valores vazios for maior que 25% excluir
        remover_colunas.append(coluna)

for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa].drop(remover_colunas, axis=1)
    fundamentos[empresa] = fundamentos[empresa].fillna(0) #preencher valores vazios com valor mais próximo (acima)
```

```python
#transformação final da base de dados
base_dados = pd.DataFrame()
for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa][1:-1]#remover primeira e última linha
    fundamentos[empresa] = fundamentos[empresa].reset_index(drop=True)
    base_dados = base_dados.append(fundamentos[empresa])
display(base_dados)
```

## 2.2 Análise de categorias

```python
correlacoes = base_dados.corr()

fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(correlacoes, cmap="Wistia", ax=ax)
plt.show()
display(correlacoes)
```

```python
correlacoes_encontradas = []
remover = []
for coluna in correlacoes:
    if correlacoes[coluna].isna().sum() ==73:
        remover.append(coluna)
        print(f"Valores NaN encontrados na coluna: {coluna}")
    for linha in correlacoes.index:
        if linha != coluna:
            valor = abs(correlacoes.loc[linha, coluna])
            if valor > 0.8 and (coluna, linha, valor) not in correlacoes_encontradas:
                correlacoes_encontradas.append((linha, coluna, valor))
                print(f"Correlação Encontrada: {linha} e {coluna}. Valor: {valor}")
```

```python
#remoção baseada na correlação + colunas com valores constantes
remover = remover + ['Imobilizado', 'Passivo Total', 'Aplicações Financeiras Avaliadas ao Custo Amortizado', 'Passivo Não Circulante', 'Empréstimos e Financiamentos', 'Provisão para IR e Contribuição Social', 'Patrimônio Líquido', 'Ajustes de Avaliação Patrimonial', 'Despesas Com Vendas', 'Despesas Financeiras','Lucro/Prejuízo do Período']
base_dados = base_dados.drop(remover, axis=1)
```

## 2.3 Feature Selection

```python
# quick - treinando uma árvore de decisão para pegar suas características mais importantes
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier(random_state=42)
x = base_dados.drop("Decisao", axis=1)
y = base_dados["Decisao"]
modelo.fit(x, y)

caracteristicas_importantes = pd.DataFrame(modelo.feature_importances_, x.columns).sort_values(by=0, ascending=False)
display(caracteristicas_importantes)
top10 = list(caracteristicas_importantes.index)#[:10]
#top10.extend(("Passivo Circulante","Investimentos","Lucros/Prejuízos Acumulados"))

print(top10)
```

## 3. Processamento de dados

## 3.1 Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

def ajustar_scaler(tabela_original):
    scaler = StandardScaler()
    tabela_auxiliar = tabela_original.drop("Decisao", axis=1)
    
    tabela_auxiliar = pd.DataFrame(scaler.fit_transform(tabela_auxiliar), tabela_auxiliar.index, tabela_auxiliar.columns)
    tabela_auxiliar["Decisao"] = tabela_original["Decisao"]
    return tabela_auxiliar

nova_base_dados = ajustar_scaler(base_dados)
top10.append("Decisao") 

nova_base_dados = nova_base_dados[top10].reset_index(drop=True)
display(nova_base_dados.head(10))
```

## 3.2 split train test

```python
from sklearn.model_selection import train_test_split

x = nova_base_dados.drop("Decisao", axis=1)
y = nova_base_dados["Decisao"]

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, random_state=42)
```

## 3.3 dummy classifier

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import classification_report, confusion_matrix

dummy = DummyClassifier(strategy="stratified", random_state=42)
dummy.fit(x_treino, y_treino)
previsao_dummy = dummy.predict(x_teste)
```

## 3.4 Avaliação

```python
def avaliar(y_teste, previsoes, nome_modelo):
    print(nome_modelo)
    report = classification_report(y_teste, previsoes)
    print(report)
    cf_matrix = pd.DataFrame(confusion_matrix(y_teste, previsoes), index=["Vender", "Comprar"], columns=["Vender", "Comprar"])
    sns.heatmap(cf_matrix, annot=True, cmap="Blues", fmt=',')
    plt.show()
    print("#" * 50)
    
    
avaliar(y_teste, previsao_dummy, "Dummy")
```

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

modelos = {
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(random_state=42),
    "ExtraTree": ExtraTreesClassifier(random_state=42),
    "GradientBoost": GradientBoostingClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "LogisticRegression": LogisticRegression(random_state=42),
    "NaiveBayes": GaussianNB(),
    "SVM": SVC(random_state=42),
    "RedeNeural": MLPClassifier(random_state=42, max_iter=400),
}
```

```python
for nome_modelo in modelos:
    modelo = modelos[nome_modelo]
    modelo.fit(x_treino, y_treino)
    previsoes = modelo.predict(x_teste)
    avaliar(y_teste, previsoes, nome_modelo)
    modelos[nome_modelo] = modelo
```

## 3.5 Tunagem

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, precision_score

modelo_final = modelos["RandomForest"] #escolhido de acordo com o maior f1-score acima

n_estimators = range(10, 251, 30)
max_features = list(range(2, 11, 2))
max_features.append('auto')
min_samples_split = range(2, 11, 2)

precision2_score = make_scorer(precision_score, labels=[0], average='macro') #label 2=compra, 0=venda

grid = GridSearchCV(
        estimator=RandomForestClassifier(),
        param_grid={
            'n_estimators': n_estimators,
            'max_features': max_features,
            'min_samples_split': min_samples_split,
            'random_state': [42],
        },
        scoring=precision2_score,
)

resultado_grid = grid.fit(x_treino, y_treino)
print("Ajuste feito")
```

```python
#avaliar o modelo tunado
modelo_tunado = resultado_grid.best_estimator_
previsoes = modelo_tunado.predict(x_teste)
avaliar(y_teste, previsoes, "RandomForest Tunado")
```
